## ラインプロット

### 書き方

コード8.2.1

```python
x = [1, 2, 3]
y = [10, 30, 20]

plt.plot(x, y)
plt.show()
```

### 曲線のプロット

コード8.2.2

```python
x = py4macro.xvalues(-4, 4, 50)
y = [i**2 for i in x]

plt.plot(x, y)
plt.show()
```

コード8.2.3

```python
def normal_dist(z):
    f = (  np.exp( -z**2 / 2 ) /
           ( 2*np.pi )**(0.5)  )
    return f

y = [normal_dist(i) for i in x]

plt.plot(x, y)
plt.show()
```

### 重ねてプロット

コード8.2.4

```python
y0 = [normal_dist(i) for i in x]
y1 = [normal_dist(i-2.5) for i in x]

plt.plot(x, y0)
plt.plot(x, y1)
plt.show()
```

コード8.2.5

```python
x = [1, 2, 3]
y0 = [10, 30, 25]
y1 = [30, 10, 15]
```

コード8.2.6

```python
plt.plot(x, y0,
         linestyle=":",
         linewidth=2,
         color="grey",
         marker="o",
         markersize=7)
plt.plot(x, y1,
         linestyle="-",
         linewidth=2,
         color="black",
         marker="^",
         markersize=7)
plt.show()
```

## ヒストグラム

コード8.3.1

```python
z0 = [random.gauss()
      for _ in range(10_000)]

plt.hist(z0)
plt.show()
```

コード8.3.2

```python
plt.hist(z0,
         bins = 20,
         linewidth=0.5,
         color="black",
         edgecolor="white",
         alpha=0.5,
         label="zの値",
         density=True)
plt.legend()
plt.show()
```

コード8.3.3

```python
z1 = [random.gauss(2, 2) for _ in range(10_000)]

plt.hist(z0,
         bins = 10,
         color="black",
         edgecolor="white",
         alpha=0.5,
         label="z0"
        )
plt.hist(z1,
         bins = 20,
         color="white",
         edgecolor="black",
         alpha=0.5,
         label="z1")
plt.title("z0とz1のヒストグラム", size=12)
plt.legend(fontsize=8)
plt.show()
```

## 散布図

コード8.4.1

```python
n = 20
x2 = [random.gauss()
      for _ in range(n)]
u2 = [random.gauss()
      for _ in range(n)]
y2 = [i + 0.5*j for i, j in zip(x2, u2)]

plt.scatter(x2, y2)
plt.show()
```

コード8.4.2

```python
x3 = [random.gauss()
      for _ in range(n)]
u3 = [random.gauss()
      for _ in range(n)]
y3 = [-i + 0.5*j
      for i, j in zip(x3, u3)]

plt.scatter(x2, y2,
            marker="o",
            s=50,
            color="grey",
            alpha=0.8,
            label="(x2,y2)")
plt.scatter(x3, y3,
            marker="x",
            s=50,
            color="black",
            label="(x3,y3)")

plt.title("散布図", size=18)
plt.xlabel("x2, x3", size=15)
plt.ylabel("y2, y3", size=15)
plt.legend(fontsize=10)
plt.show()
```
