```python
c_lst = []

for y in y_lst:
    c = 0.1 + 0.6 * y
    c_lst.append(c)

c_lst
```

```python
def consumption(lst):

    c_lst = []

    for y in lst:
        c = 0.1 + 0.6 * y
        c_lst.append(c)

    return c_lst
```

### 例5：range()（資産の推移）

コード6.3.8

```python
wealth = 100
wealth_lst = [wealth]

for _ in range(5):
    wealth *= (1 + 0.03)
    wealth_lst.append(wealth)

wealth_lst
```

コード6.3.9

```python
wealth_lst = []

for t in range(6):
    wealth = 100 * (1+0.03)**t
    wealth_lst.append(wealth)
```

## whileループ

### whileループの書き方

### 例1：簡単なパターン

コード6.5.1

```python
components = ["消費", "投資", "政府支出", "純輸出"]
i = 0

while i < len(components):
    print(components[i])
    i += 1
```

### 例2：目標資産額に必要な期間

コード6.5.2

```python
wealth = 100
target = 200
r = 0.03
year = 0

while wealth < target:
    wealth *= 1 + r
    year += 1

print(f"{year}年かかります。")
```

コード6.5.3

```python
def wealth_years(wealth=100,
                 target=200,
                 r=0.03):
    """
    引数：
        wealth: 初期資産
        target: 目標資産額
        r: 年間収益率
    戻り値：
        目標資産額以上の達成に
        必要な年数（表示あり）
    """
    year = 0

    while wealth < target:
        wealth *= 1 + r
        year += 1

    print(f"必要年数：{year}")
    return year
```

### 例3：壁までの距離

コード6.5.4

```python
distance = 1
count = 0

while distance > 0.001:

    # print(distance)
    distance -= distance/2
    count += 1

print(f"ループの回数：{count}")
print(f"残りの距離：{distance*1000}mm")
```
